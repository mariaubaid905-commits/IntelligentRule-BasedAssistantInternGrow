```python

import requests

from bs4 import BeautifulSoup
```

```python
# Define a dictionary containing pre-mapped rules and corresponding responses
PRE_MAPPED_RESPONSES = {
    "hello": "Hello! How can I assist you today?",
    "hi": "Hi there! What would you like to know?",
    "how are you": "I am an intelligent assistant. I am functioning optimally!",
    "what is your name": "I am the InternGrow Rule-Based Assistant.",
    "bye": "Goodbye! Have a great day.",
    "who created you": "I was developed as part of the InternGrow Python Programming Track."
}

def get_premapped_response(user_input):
    """
    Evaluates the user's input against predefined rules.
    Returns the corresponding response if a match is found.
    """
    # Clean the input by converting it to lowercase and removing leading/trailing spaces
    cleaned_input = user_input.strip().lower()
    
    # Check if the cleaned input exists in our knowledge base
    if cleaned_input in PRE_MAPPED_RESPONSES:
        return PRE_MAPPED_RESPONSES[cleaned_input]
    
    # Return None if no pre-mapped response is found
    return None
```

```python
def scrape_wikipedia_summary(query):
    """
    Scrapes a real-time summary from Wikipedia for unknown queries
    using requests and BeautifulSoup. Includes User-Agent to bypass blocks.
    """
    # Convert query to Title Case (e.g., 'quantum computing' -> 'Quantum_Computing')
    formatted_query = query.strip().title().replace(" ", "_")
    url = f"https://en.wikipedia.org/wiki/{formatted_query}"
    
    # Wikipedia strictly requires a User-Agent header to allow scraping
    headers = {
        'User-Agent': 'InternGrow_AssistantBot/1.0 (Learning Project)'
    }
    
    try:
        # Send an HTTP GET request with the defined headers
        response = requests.get(url, headers=headers)
        
        # Verify that the web page exists (Status Code 200)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            
            # Extract all paragraph tags
            paragraphs = soup.find_all('p')
            
            # Iterate through paragraphs to find the first substantial text block
            for p in paragraphs:
                text = p.get_text().strip()
                if len(text) > 60:
                    # Return the first 3 sentences to keep the summary concise
                    sentences = text.split('.')
                    summary = '.'.join(sentences[:3]) + '.'
                    return f"Wikipedia Summary: {summary}"
                    
        # Improved error handling to show exactly why it failed
        return f"I'm sorry, I couldn't find a specific Wikipedia summary for that query. (Error Code: {response.status_code})"
        
    except Exception as e:
        return f"An error occurred while attempting to scrape data: {e}"
```

```python
def run_assistant():
    """
    Initializes the intelligent assistant and manages the continuous user interaction loop.
    """
    print("🤖 Intelligent Rule-Based Assistant Initiated.")
    print("Type your message below. Type 'quit' or 'exit' to terminate the session.\n")
    print("-" * 50)
    
    # Maintain a continuous loop for ongoing conversation
    while True:
        # Capture user input
        user_input = input("You: ")
        
        # Define termination conditions for the loop
        if user_input.strip().lower() in ['quit', 'exit']:
            print("Bot: Terminating session. Goodbye!")
            break
            
        # Prevent processing of empty strings
        if not user_input.strip():
            print("Bot: Please enter a valid prompt.")
            continue
            
        # Step 1: Attempt to retrieve a pre-mapped response
        response = get_premapped_response(user_input)
        
        # Step 2: Fall back to real-time web scraping if no mapped response exists
        if not response:
            print(f"Bot: Let me search the web for '{user_input}'...")
            response = scrape_wikipedia_summary(user_input)
            
        # Display the final generated response
        print(f"Bot: {response}\n")

# Execute the assistant
run_assistant()
```
